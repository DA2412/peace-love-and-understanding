# Weapons4Ukraine

Here is a brief analysis of the weapon systems that have been provided from NATO members to Ukraine from 2014 to 2021. These include the *orders* and the effective *delivered* weapon systems for both Ukraine and "rebels" of Donbass (provided by Russia). The data are available at [SIPRI Arms Transfers Database](https://sipri.org/databases/armstransfers) and the last update is to year 2021. 

Here are the ordered weapon systems per year and highlighted by supplier country. 

```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
import pandas as pd
import numpy as np
sns.set_style("darkgrid")
sns.set(font_scale = 0.8)
inFile = "NATOUkraineCSV.csv"
df = pd.read_csv(inFile,sep=";")

fig,ax = plt.subplots(figsize=(6,3))
sns.barplot(data=df,x="Supplier", y = "no_ordered", hue="year_order",
            ci=None,palette='Blues',ax=ax)
ax.legend(title='year of demand',loc='upper center')
plt.xlabel(" ")
plt.ylabel("Number of ordered weapon systems")
plt.title('2014-2021')
plt.tight_layout()
```

The ordered weapon system have been all supplied within a year; the only exception is the 2021 which would be probably be supplied in 2022.

```python
fig,ax = plt.subplots(figsize=(6,3))
sns.barplot(data=df,x="Supplier", y = "no_delivered", hue="year_delivered",
            ci=None,palette='Reds',ax=ax)
ax.legend(title='year of supply',loc='center')
plt.xlabel(" ")
plt.ylabel("Number of delivered weapon systems")
plt.title('2014-2021')
plt.tight_layout()
```

Let us have a look to the different typology of delivered weapon systems.

Russia sent to Donbass Republics a total of 125 weapon systems, particularily:
* 10 second hand surface-to-air missiles (SAM)
* 100 anti-tank missiles (Kornet)
* 10 Second-hand but modernized systems included tanks (T-72B modernized to T-72B3)
* 5 Infantry fighting vehicles (IFV) like BTR-82A.

NATO members sent the following weapon systems:
* 87 Infantry fighting vehicles (IFV) (BMP-1)
* 390 anti-tank missile (FGM-148 Javelin) 
* 5 patrol craft (FPB-98 Mk-1 version): to be delivered in 2022-2024, EUR137m deal (incl 5 years support; 85% financed with French loans); for border guard; including 5 produced under licence in Ukraine
* 56 self-propelled guns (2S1 122mm)
* 4 transport helicopters (EC725 Super Cougar) 
* 1 trainer/combat aircraft (L-39Z Albatros)
* 129 Armoured personnel carrier (APC): MT-LB
* 50 Armoured patrol vehicle (APV): Dozor-B
* 24 APC/APV
* 12 Armed unmanned aerial vehicle (UAV) also known as drone (Bayraktar TB-2)
* 15 arty locating radar (TPQ-36 Firefinder)

```python
dfp = df.dropna()
ind_nato = ~dfp['Supplier'].isin(["Russia"])
dfp.loc[ind_nato,"Supplier"] = 'NATO'
dfp_tot = dfp.groupby(['info_weapon','Supplier'])['no_delivered'].sum().reset_index()
fig,ax = plt.subplots(figsize=(5,3))
sns.barplot(data=dfp_tot,x="no_delivered", y = "info_weapon", hue="Supplier", ci=None,ax=ax)
ax.legend(title='Supplier',loc='lower right')
plt.xlabel("Number of delivered weapon systems")
plt.ylabel("Designation of weapon system")
plt.tight_layout()
plt.title('2014-2021')
```

In summary, NATO sent 773 weapon systems in the time window 2014-2021; in the same period of time Russia sent 125 weapons systems.

```python
df_nato_tot = dfp.groupby(['Supplier'])['no_delivered'].sum().reset_index()
fig,ax = plt.subplots(figsize=(2,4))
sns.barplot(data=df_nato_tot,x= "Supplier", y ="no_delivered", ci=None,ax=ax)
plt.ylabel("Number of delivered weapon systems")
plt.xlabel("Supplier")
plt.title('2014-2021')
plt.tight_layout()
```
